```python
#Importing Initial Libraries
import numpy as np
import pandas as pd
```

```python
#Importing Excel File
df = pd.read_excel('BengaluruHouseData.xlsx')
df.head()
```

```python
df.groupby('area_type')['area_type'].agg('count')
```

```python
df.shape
```

```python
#Droping unwanted columns
df = df.drop(['society','availability'], axis=1)
df.head()
```

```python
df.isnull().sum()
```

```python
#Droping rows where location or size column is None
df = df.dropna(subset=['location','size'], axis=0)
df.shape
```

```python
#Filling None values of bath column with median of bath column
df['bath'].fillna(df['bath'].median(), inplace=True)
df.isnull().sum()
```

```python
#Filling None values of balcony column with 0
df['balcony'].fillna(0, inplace=True)
df.describe()
```

```python
df['size'].unique()
```

```python
#cleaning size column and creating new bhk column
df['bhk'] = df['size'].apply(lambda x: int(x.split(' ')[0]))
df.head()
```

```python
#Drop size column
df.drop('size', axis=1, inplace=True)
df.head()
```

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
df[~df['total_sqft'].apply(is_float)]
```

```python
# Defining a function to convert range value into average of range
def convert_sqft_to_num(x):
    if type(x) == str:
        token = x.split('-')
        if len(token) == 2:
            return (float(token[0])+float(token[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
df['total_sqft'] = df['total_sqft'].apply(convert_sqft_to_num)
```

```python
df['price_per_sqft'] = round(df['price']*100000/df['total_sqft'], 0)
df.head()
```

```python
#location doesn't have extra space
df['location'] = df['location'].apply(lambda x: x.strip())
```

```python
location_stats = df.groupby('location')['location'].agg('count').sort_values(ascending=False)
location_stats
```

```python
len(location_stats[location_stats<=10])
```

```python
location_stats_less_than_10 = location_stats[location_stats<=10]
location_stats_less_than_10
```

```python
df['location'] = df['location'].apply(lambda x : 'other' if x in location_stats_less_than_10 else x)
df.head()
```

```python
#Data Errors
df[df['total_sqft']/df['bhk']<300].head()
```

```python
df = df[~(df['total_sqft']/df['bhk']<300)]
```

```python
def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf['price_per_sqft'])
        sd = np.std(subdf['price_per_sqft'])
        reduced_df = subdf[(subdf['price_per_sqft']>(m-sd)) & (subdf['price_per_sqft']<(m+sd))]
        df_out = pd.concat([df_out, reduced_df], ignore_index=True)
    return df_out
```

```python
df = remove_pps_outliers(df)
df.shape
```

```python
#we need to clean the outliers because we know 2bhk can't be more expensive than 3bhk in same location

def remove_bhk_outliers(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df['price_per_sqft']),
                'std': np.std(bhk_df['price_per_sqft']),
                'count': bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df['price_per_sqft']<(stats['mean'])].index.values)
    return df.drop(exclude_indices,axis=0)
```

```python
df = remove_bhk_outliers(df)
df.shape
```

```python
df[df['bath'] > (df['bhk']+2)]
```

```python
df = df[~(df['bath']>(df['bhk']+2))]
df.shape
```
